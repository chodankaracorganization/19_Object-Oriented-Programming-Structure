Problem Statement 1: Create a Constructor for a Person

Write a JavaScript function constructor named Person that takes two parameters, name, and age, and Add a
method to the prototype to display name and age. Then, create two instances of Person and display their
names and ages.

```python
function Person(name, age) {
  this.name = name;
  this.age = age;
}

Person.prototype.display = function () {
  return `${this.name}'s age is ${this.age}`;
};

const person1 = new Person("Ashish", 30);
const person2 = new Person("John", 32);

console.log(person1.display());
console.log(person2.display());
```

Problem Statement 2: Implement a Bank Account

Create a constructor function named BankAccount that initializes a bank account with an initial balance.
Include methods for depositing and withdrawing money from the account. Write code to demonstrate these
operations on a bank account object.


```python
// BankAccount constructor
function BankAccount(initialBalance) {
    this.balance = initialBalance;

    // Method to deposit money
    this.deposit = function(amount) {
        if (amount > 0) {
            this.balance += amount;
            console.log(`Deposited $${amount}. New balance: $${this.balance}`);
        } else {
            console.log("Deposit amount must be positive");
        }
    };

    // Method to withdraw money
    this.withdraw = function(amount) {
        if (amount > 0) {
            if (amount <= this.balance) {
                this.balance -= amount;
                console.log(`Withdrew $${amount}. New balance: $${this.balance}`);
            } else {
                console.log("Insufficient funds");
            }
        } else {
            console.log("Withdrawal amount must be positive");
        }
    };

    // Method to check balance
    this.checkBalance = function() {
        console.log(`Current balance: $${this.balance}`);
        return this.balance;
    };
}

// Creating a bank account instance
const myAccount = new BankAccount(1000);

// Demonstrating operations
myAccount.checkBalance();  // Current balance: $1000
myAccount.deposit(500);    // Deposited $500. New balance: $1500
myAccount.withdraw(200);   // Withdrew $200. New balance: $1300
myAccount.withdraw(1500);  // Insufficient funds
myAccount.deposit(-50);    // Deposit amount must be positive
myAccount.checkBalance();  // Current balance: $1300
```

Problem Statement 5: Implement a Calculator Class

Create a class called Calculator that initializes two values value1 and value2 to store numbers. Add methods for
add, subtract, multiply, and divide. Perform and show operations.

```python
class Calculator {
  value1;
  value2;

  constructor(value1, value2) {
    this.value1 = value1;
    this.value2 = value2;
  }

  add() {
    const result = this.value1 + this.value2;
    console.log(`${this.value1} + ${this.value2} = ${result}`);
    return result;
  }

  subStract() {
    const result = this.value1 - this.value2;
    console.log(`${this.value1} - ${this.value2} = ${result}`);
    return result;
  }

  multiPly() {
    const result = this.value1 * this.value2;
    console.log(`${this.value1} * ${this.value2} = ${result}`);
    return result;
  }

  divide() {
    if (this.value2 === 0) {
      console.log("Error:Division by zero is not allowed");
      return undefined;
    }
    const result = this.value1 / this.value2;
    console.log(`${this.value1} / ${this.value2} = ${result}`);
    return result;
  }
}

const calc = new Calculator(10, 5);

calc.add();
calc.subStract();
calc.multiPly();
calc.divide();
```

Problem Statement 6: Design a Class for a Geometric Shape

Design a class called Shape that can represent various geometric shapes. Implement subclasses for specific
shapes like Circle, Rectangle, and RightTriangle. Each subclass should have properties of that shape (e.g..
Width, height for a rectangle) and methods for calculating the area and perimeter of the shape. Create
instances of these shapes and calculate their areas and perimeters.


```python
class Shape {}

class Circle extends Shape {
  constructor(radius) {
    super();
    this.radius = radius;
  }

  area() {
    return Math.PI * Math.pow(this.radius, 2);
  }

  perimeter() {
    return 2 * Math.PI * this.radius;
  }

  display() {
    console.log("Circle Properties");
    console.log(`Radius:${this.radius}`);
    console.log(`Area:${this.area().toFixed(2)}`);
    console.log(`Circumference:${this.perimeter().toFixed(2)}`);
  }
}

class Rectangle extends Shape {
  constructor(width, height) {
    super();
    this.width = width;
    this.height = height;
  }

  area() {
    return this.width * this.height;
  }
  perimeter() {
    return 2 * (this.width + this.height);
  }

  display() {
    console.log("Rectangle Properties");
    console.log(`Width:${this.width},Height:${this.height}`);
    console.log(`Area:${this.area().toFixed(2)}`);
    console.log(`Perimeter:${this.perimeter().toFixed(2)}`);
  }
}

class RightTriangle extends Shape {
  constructor(base, height) {
    super();
    this.base = base;
    this.height = height;
  }

  area() {
    return (this.base * this.height) / 2;
  }

  perimeter() {
    const hypotenuse = Math.sqrt(
      Math.pow(this.base, 2) + Math.pow(this.height, 2)
    );
    return this.base + this.height + hypotenuse;
  }

  display() {
    console.log("RightTriangle Properties");
    console.log(`Base:${this.base},Height:${this.height}`);
    console.log(`Area:${this.area().toFixed(2)}`);
    console.log(`Perimeter:${this.perimeter().toFixed(2)}`);
  }
}

const circle = new Circle(5);
const rectangle = new Rectangle(5, 10);
const rightTriangle = new RightTriangle(10, 20);

circle.display();
console.log("----------------");
rectangle.display();
console.log("----------------");
rightTriangle.display();
```

Problem Statement 7: Implement a Library System

Create a class called Library that can manage a collection of books. Each book should be represented as an
instance of a Book class (with properties like title, author, and availability). Implement methods in the Library
class to add books, check out books, and return books. Keep track of the books available.

```python
class Book {
  constructor(title, author) {
    this.title = title;
    this.author = author;
    this.available = true; // Default to available
    this.dueDate = null;
  }

  getInfo() {
    return `${this.title} by ${this.author} - ${
      this.available ? 'Available' : `Checked out (Due: ${this.dueDate})`
    }`;
  }
}

class Library {
  constructor() {
    this.books = []; // Collection to store books
  }

  // Add a new book to the library
  addBook(title, author) {
    const newBook = new Book(title, author);
    this.books.push(newBook);
    console.log(`Added: ${newBook.getInfo()}`);
    return newBook;
  }

  // Find a book by title (case-insensitive)
  findBook(title) {
    return this.books.find(
      book => book.title.toLowerCase() === title.toLowerCase()
    );
  }

  // Check out a book
  checkoutBook(title) {
    const book = this.findBook(title);

    if (!book) {
      console.log(`Book not found: "${title}"`);
      return false;
    }

    if (!book.available) {
      console.log(`"${title}" is already checked out`);
      return false;
    }

    // Set due date 14 days from now
    const dueDate = new Date();
    dueDate.setDate(dueDate.getDate() + 14);

    book.available = false;
    book.dueDate = dueDate.toLocaleDateString();

    console.log(`Checked out: "${title}". Due on ${book.dueDate}`);
    return true;
  }

  // Return a book
  returnBook(title) {
    const book = this.findBook(title);

    if (!book) {
      console.log(`Book not found: "${title}"`);
      return false;
    }

    if (book.available) {
      console.log(`"${title}" was not checked out`);
      return false;
    }

    book.available = true;
    book.dueDate = null;

    console.log(`Returned: "${title}"`);
    return true;
  }

  // List all books
  listBooks() {
    console.log('\nLibrary Collection:');
    if (this.books.length === 0) {
      console.log('No books in collection');
      return;
    }

    this.books.forEach(book => {
      console.log(book.getInfo());
    });
  }
}

// Example Usage
const myLibrary = new Library();

// Add books
myLibrary.addBook('The Hobbit', 'J.R.R. Tolkien');
myLibrary.addBook('1984', 'George Orwell');
myLibrary.addBook('To Kill a Mockingbird', 'Harper Lee');

// Checkout and return books
myLibrary.checkoutBook('The Hobbit');
myLibrary.checkoutBook('1984');
myLibrary.returnBook('1984');

// List all books
myLibrary.listBooks();
```
